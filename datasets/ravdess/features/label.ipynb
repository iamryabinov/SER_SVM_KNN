```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv('features_with_labels.csv', delimiter=';')
features
```

```python
features['label'] = [inst.split('-')[2] for inst in features['name'].values.tolist()]
features
```

```python
features.groupby(['label']).count()
```

```python
features.loc[features['label'] == '05', 'label'] = 'ang'
features.loc[features['label'] == '07', 'label'] = 'dis'
features.loc[features['label'] == '06', 'label'] = 'fea'
features.loc[features['label'] == '03', 'label'] = 'hap'
features.loc[features['label'] == '01', 'label'] = 'neu'
features.loc[features['label'] == '04', 'label'] = 'sad'
features.loc[features['label'] == '02', 'label'] = 'cal'
features.loc[features['label'] == '08', 'label'] = 'sur'
features
```

```python
features.groupby(['label']).count()
```

```python
features.to_csv('features_with_labels.csv', index=False, sep=';')
```
